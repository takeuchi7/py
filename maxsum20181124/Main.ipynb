```python
#coding:utf-8
import sys, json
from argparser import ArgParser
from factor_graph import FactorGraph
from scheduler import Scheduler
from variableagent import VariableAgent
from functionagent import FunctionAgent
from messageserver import MessageServer

opt_pattern = {'-l': {'name': 'lambda', 'type': 'int', 'default': 10}}
#'-l'
#'name': 'lambda'
#'type': 'int'
#'default': 10...可変エージェントの収束を保証するために_z_ベクトルが変更されない回数?
#花房先生いわく、これをオプションの代わりにして導入しているということ。
#つまり、やはり実行はターミナルでpython ms.py fg.jasonで良いということであるな

arg = ArgParser(sys.argv[2:], opt_pattern)#ここクラスを代入し再定義している
#sys.argvは３つのリスト 
##sys.argv[0]...ipykernel_launcher.py(Anaconda3からのディレクトリ)
##sys.argv[1]...-f(なにこれ？指定のないファイル)
##sys.argv[2]...kernel-780bfd0a-9d79-43c9-969c-1087e48bd09c.json(AppDataからのディレクトリ)
#なるほどsys.argvは入力をいくらでも受け取れるものなのか
#ああ！だからargv[2:]なのか。おそらく複数入力することを想定しているのか。しかしだからこそargv[1]に何を入れるべきなのか
#もしかしてargv[1]にはfg.jasonを入れて置き、それ以降は自分で考えて作るものなのか？？？
#もし入力のリストと被っているところがあれば、argに書き出す

opt = arg.read()
#optに入力したものを書き出している。入力の処理段階でアルゴリズムの本質ではなさそう
#ターミナルでpython ms.py fg.json 20 と入力したら　lambda:20になった
#ここでlambda :100 にしたら処理が重たくなった
#なるほどね？おそらくパラメータの入力をややこしくしているということか
#opt_pattern = {'-l': {'name': 'lambda', 'type': 'int', 'default': 10}}であれば、ターミナルでpython ms.py fg.json 20（int型）
#opt_pattern = {'-l': {'name': 'lambda', 'type': 'bool', 'default': 10}}であれば、ターミナルでpython ms.py fg.json True（bool型）
#といった感じで、入力の型をopt_patternで指定して、指定通りの型が入力されれば、従うということか
#はぁーわかったけどめんどくさ！！！

fg = FactorGraph(opt)
#因子グラフ

#fg.load(sys.argv[1])
fg.load('fg.json')#こうすれば万事解決じゃね？いや、jsonでエラーが出た/''を付けたら治った。。。
#インスタンスを定義した時点でこいつは_init_をもっている
#A = json.load(open('fg.json', 'r'))
#print(A)
#fg.load(A)
#fg.variables fg.functions)

#self.variables = {}
#self.functions = {}
#self.constants = {}をloadでいれている

#つまり上の関数より現時点では以下の関数を持つことになる
#self.variables = {v0,v1,v2,v3,v4}
   #self.variables = {Variable(v0,[-1~1], [f1], ['v2', 'v1']),Variable(v1,[-1~1], [f1], ['v0', 'v2']),...,v4}
#self.functions = {f1,f2}
#self.constants = {}

#function(f1)=Function(f1,[v0,v1,v2],self)
#function(f2)=Function(f2,[v0,v1,v2],self)
#self.variables[v0] = Variable(v0,[-1~1], [f1], ['v2', 'v1'])
#self.variables[v1] = Variable(v1,[-1~1], [f1], ['v0', 'v2'])
#self.variables[v2] = Variable(v2,[-1~1], [f1,f2], ['v4', 'v0', 'v1', 'v3'])
#self.variables[v3] = Variable(v3,[-1~1], [f2], ['v4', 'v2'])
#self.variables[v4] = Variable(v4,[-1~1], [f2], ['v2', 'v3'])

ms = MessageServer(opt)
#ほほう、msにopt{lambda:10}を渡していると
#ms.load
#ms.send
#ms.printer
#ms.logger
agents = {}

for v in fg.variables:
#{v0:,v1:,v2:,v3:,v4:}
#各変数に対してクラスを渡している。。。下準備やべぇ
	agent = VariableAgent(v, fg, ms, opt)
	agents[v] = agent
#agents={v0:関数 v1:関数 v2:関数 v3:関数 v4:関数}

for f in fg.functions:
#{f1:,f2:}
#各因子に対してクラスを渡している。。。下準備やべぇ
	agent = FunctionAgent(f, fg, ms, opt)
	agents[f] = agent
#agents={v0:関数 v1:関数 v2:関数 v3:関数 v4:関数 f1:関数 f2:関数}

#すげー！！クラスを定義したやつをさらに引数にしている！
#scheduler = Scheduler(FactorGraph(opt), {v0:関数 v1:関数 v2:関数 v3:関数 v4:関数 f1:関数 f2:関数}, ms, {lambda:10})
scheduler = Scheduler(fg, agents, ms, opt)

scheduler.initialize()
#各変数ノードと因子ノードにつき
#self.agentLog[c] = 0
#self.agentLogTest[c] = 0
#self.timeLog[c] = 0
scheduler.run()
scheduler.terminate()
```
